¿Con que test lo comprobarías?

```r
airbnb_E<-airbnb %>%
    group_by(TipoAlquiler) %>%
    filter ( TipoAlquiler=='Entire home/apt')

p_m_Entrie_home<-mean(airbnb_E$Precio, na.rm=TRUE)
print(paste0("La media del precio para el tipo Entire home/apt: ",p_m_Entrie_home))

P_Private_room<-airbnb %>%
    group_by(TipoAlquiler) %>%
    filter ( TipoAlquiler=='Private room')

P_m_Private_room<-mean(P_Private_room$Precio, na.rm=TRUE)

print(paste0("La media del precio para el tipo Private room: ",P_m_Private_room))
P_Shared <-airbnb %>%
    group_by(TipoAlquiler) %>%
    filter ( TipoAlquiler=='Shared room')

p_m_Shared<-mean(P_Shared$Precio, na.rm=TRUE)
print(paste0("La media del precio para el tipo Shared room: ",p_m_Shared))
```

Una vez que hayamos filtrado los datos correspondientes calcular el valor o la combinación de valores que mejor nos permite obtener el precio de un inmueble.

```r
model<-lm(data=barrio_sol, formula = Precio ~ MetrosCuadrados+NumDormitorios+NumBanyos+MaxOcupantes)
summary(model)
confint(model)
```
